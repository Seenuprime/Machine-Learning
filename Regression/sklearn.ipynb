```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("Data.csv")
X = df.iloc[:, :-1] #independent variable
y = df.iloc[:, -1] #dependent variable
```

```python
#used to take care of the missing values (nan)
imputer = SimpleImputer(missing_values=np.nan, strategy="mean") #making strategy
imputer.fit(X.iloc[:, 1:]) #telling where to apply'''include only numerical value columns'''
X.iloc[:, 1:] = imputer.transform(X.iloc[:, 1:]) #returns the currented / filled version, so store it
```

```python
X
```

One Hot Encoding:(independent variable)
it is used to transform string data to numerical data to pass that to a model

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

# Lableing the dependent variables

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```
