```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Position_Salaries.csv")
X = df.iloc[:, 1:-1].values
y = df.iloc[:, -1].values
```

Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
lin_regressor = LinearRegression()
lin_regressor.fit(X, y)
y_pre = lin_regressor.predict(X)
```

Polynomial Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X) #converting x to x^1, x^2, x^3, ect...
```

```python
lin_regressor_2 = LinearRegression()
lin_regressor_2.fit(X_poly, y)
y_pre_2 = lin_regressor_2.predict(X_poly)
```

Visualizing the Linear Model

```python
plt.scatter(X, y, color="red")
plt.plot(X, y_pre, color="blue")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Linear Model")
plt.show()
```

Visualising Polynomial Model

```python
plt.scatter(X, y, color="red")
plt.plot(X, y_pre_2, color="blue")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Polinomial Model")
plt.show()
```

Predicting new value with Linear Model

```python
lin_regressor.predict([[3], [4]])
```

Predicting new value with Polynomial model

```python
lin_regressor_2.predict(poly_reg.fit_transform([[6]]))
```
