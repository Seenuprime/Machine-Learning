```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Position_Salaries.csv")
X = df.iloc[:, 1:-1]
y = df.iloc[:, -1]
```

```python
from sklearn.ensemble import RandomForestRegressor
RF_regressor = RandomForestRegressor(n_estimators = 10, random_state=0) #n_estimator = 10 means 10 trees
RF_regressor.fit(X, y)
```

```python
RF_regressor.predict([[1.5], [2.5], [3.6], [4.8], [5.3], [6.6], [7], [8.4], [9], [10.4]])
```

```python
y_pre = RF_regressor.predict(X)
plt.scatter(X, y, color="red")
plt.plot(X, y_pre, color="blue")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Random Forest Regressor")
plt.show()
```
