```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Data.csv")
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
y_pre = np.array(regressor.predict(X_train))
```

```python
new_y_pre = np.array(regressor.predict(X_test))
```

```python
store = np.concatenate((new_y_pre.reshape(len(new_y_pre), 1), y_test.reshape(len(y_test), 1)), axis=1)
```
